```
# maximum of three numbers
a=20
b=45
c=55
def maximum(a,b,c):
  if(a>=b)and(a>=c):
    maximum=a
  elif(b>=a)and(b>=c):
      maximum=b
  else:
        maximum=c

  return maximum
print(maximum(a,b,c))
```

```
# factorial of a number
def fact(num):
  if(num==1):
    return 1
  else:
      return num*fact(num-1)
num=4
print("factorial of ",num,"is",fact(num))
```
